```python
#Import matplotlib
import matplotlib.pyplot as plt
```

```python
#Line plot.
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
plt.plot(x, y)

plt.xlabel('This is X-Label')
plt.ylabel('This is Y-Label')
plt.title('This is our first Line Plot')

plt.show()
```

```python
plt.figure(figsize=(6,4))

x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]


#line 1 plot
plt.plot(x, y1, label = "Line 1", color="pink")
plt.plot(x, y2, label = "Line 2", color="red")

plt.grid()
plt.legend()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,25,3,4,5]
plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,4)
plt.plot(x,y2)
```

```python
x=[1,2,8,4,5]
y=[2,5,6,9,10]
x1=[1,4,7,2,5]
y2=[3,6,9,8,5]
```

```python
fig=plt.figure(figsize=(6,4))
axis1=fig.add_axes([0.1,0.1,0.8,0.8])#[left,bottom,width,height]
axis1.plot(x,y)
plt.title("First Plot")

axis2=fig.add_axes([0.2,0.5,0.2,0.2])
axis2.plot(y2,x1)
plt.title("Second Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,5,6,9,10]

plt.figure(figsize=(6,4),dpi=100)
plt.plot(x,y)
plt.grid()
```

```python
x=[1,2,3,4,5]
y1=[2,5,6,9,10]
y2=[1,2,3,4,5]

plt.plot(x,y1,label= "Nili line",color='black',lw=2.5,ls=':',marker='+')
plt.plot(x,y2,label= "lal line",color='red',lw=2.5,ls='-.',marker="*")

plt.grid()
plt.legend()
```
